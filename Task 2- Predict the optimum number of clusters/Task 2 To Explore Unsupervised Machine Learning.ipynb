#  Importing Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
sns.set()

from sklearn.cluster import KMeans
```

# Loading Data

```python
data = datasets.load_iris()
data = pd.DataFrame(data.data, columns = data.feature_names)
data.head()
```

```python
x =data.iloc[:, [0, 1, 2, 3]].values
```

```python
inertia=[]
k_value= range(1,11)

for i in k_value:
    kmean=KMeans(n_clusters=i,max_iter = 300,random_state = 0)
    kmean.fit(x)
    inertia.append(kmean.inertia_)

plt.plot(k_value,inertia,marker='o')
plt.title('Elbow Plot')
plt.xlabel('K_Values or No. of Cluster')
plt.ylabel('Inertia');
```

We Choose three(3) Clusters

```python
kmean=KMeans(n_clusters=3,max_iter = 300,random_state = 0)
kmean.fit(x)
```

```python
y_pred= kmean.predict(x)
```

Visualising the clusters and cluster centers

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red',label='Iris-Setosa')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='blue',label='Iris-Versicolor')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='green',label='Iris-Virginica')
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],s=100,c='yellow',label='Cluster centers')
plt.title('Clusters')
plt.legend();
```
